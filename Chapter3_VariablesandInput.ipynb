```python
# --- 1 & 2. WHAT IS A VARIABLE & DECLARATION ---
# A variable is a container for storing data. 
# Declaration happens the moment you first assign a value.
score = 100
name = "Gemini"


# --- 3. VARIABLE ASSIGNMENT AND REFERENCE ---
# Assignment: Using '=' to put a value in the box
x = 5 
y = 10

# Reference: Using the variable name to get the value out
total = x + y
print(total) # References 'total' to show 15


# --- 4. COMPOUND ASSIGNMENT OPERATORS ---
# Shortcuts for doing math on the same variable
count = 10

count += 5  # Same as: count = count + 5 (Result: 15)
count -= 2  # Same as: count = count - 2 (Result: 13)
count *= 2  # Same as: count = count * 2 (Result: 26)
count /= 2  # Same as: count = count / 2 (Result: 13.0)

print(count)


# --- 5. THE INPUT() FUNCTION ---
# Used to get information from the user (always returns a String)
user_name = input("Enter your name: ")

# To use input as a number, you must convert it (Casting)
user_age = input("Enter your age: ")
age_next_year = int(user_age) + 1

print("Hello " + user_name)
print("Next year you will be " + str(age_next_year))
```
