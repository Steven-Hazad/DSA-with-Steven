```python
from pprint import pprint

family_tree = {
    "Vuth": ["Makara"],
    "Nita": ["Makara"],
    "Sokha": ["Rachana"],
    "Rotana": ["Rachana"],
    "Makara": ["Sothea"],
    "Rachana": ["Sothea"]
}

pprint(family_tree)

```

```python

def find_parents(child_name):
    parents = []
    for parent, children in family_tree.items():
        if child_name in children:
            parents.append(parent)
    return parents
```
