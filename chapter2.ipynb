```python
# --- 1. NUMERIC DATATYPES & ARITHMETIC ---
integer_num = 10
float_num = 3.0

print(f"Standard Division: 10 / 3 = {10 / 3}")    # Returns float (3.333...)
print(f"Floor Division:    10 // 3 = {10 // 3}")   # Rounds down (3)
print(f"Modulo (Remainder): 10 % 3 = {10 % 3}")    # Remainder (1)
print(f"Exponent (Power):   2 ** 3 = {2 ** 3}")    # 2 to the power of 3 (8)


# --- 2. OPERATOR PRIORITIES (PEMDAS) ---
# Parentheses -> Exponents -> Mult/Div -> Add/Sub
result = (2 + 3) * 4 ** 2 / 2
# 1. (2+5) = 5
# 2. 4**2 = 16
# 3. 5 * 16 / 2 = 40.0
print(f"Result of complex math: {result}")


# --- 3. DATA TYPE CONVERSION (CASTING) ---
age_str = "25"
age_int = int(age_str)         # String to Int
price_float = float("19.99")    # String to Float
score_text = str(98.5)          # Float to String

print(f"Type of age_int: {type(age_int)}")
print("Combined String: " + score_text + "% Score")


# --- 4. STRING APPLICATIONS ---
greeting = "Hello"
subject = "Python"

# Concatenation & Repetition
full_sentence = greeting + " " + subject + "!"
print(full_sentence)
print("-" * 20) # Prints 20 dashes

# Indexing (Accessing characters)
word = "CODING"
print(f"First letter: {word[0]}")  # C
print(f"Last letter:  {word[-1]}") # G

# Multi-line strings
lesson_plan = """
Chapter 1:
- Numbers
- Math
- Strings
"""
print(lesson_plan)
```
